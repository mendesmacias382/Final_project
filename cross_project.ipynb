```python
import seaborn as sns
corr=np.abs(cross_slb.corr())
#Set up mask for triangle representation
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(10, 10))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask,  vmax=1,square=True, linewidths=.5, cbar_kws={"shrink": .5},annot = corr)
plt.show()
```

```python
### 1.SET HYPOTHESIS

#  HO:  without_mark_goal = marked_goal
#  H1:  without_mark_goal != marked_goal



#  HO:  without_mark_strike = mark_strike
#  H1:  without_mark_strike != mark_strike


#  HO:  without_mark_on_target = mark_on_target
#  H1:  without_mark_on_target != mark_on_target
```

```python
data_slb = cross_slb[['marked', 'goal']]

data_slb
```

```python
without_mark_slb = data_slb[data_slb['marked'] == 0]
without_mark_slb
```

```python
marked_slb = data_slb[data_slb['marked'] ==1 ]
marked_slb
```

```python

#  HO:  without_mark_goal = marked_goal
#  H1:  without_mark_goal != marked_goal


alpha = 0.05

st.ttest_ind(marked_slb['goal'],without_mark_slb['goal'])
```
